```
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data=pd.read_csv('penguins_size.csv')# read the file
```

```
data
```

```
data.isna().sum()# check missing values
```

```
data1=data.fillna(method='ffill')# preprocessing method to fill missing values
data1
```

```
data2=data1.copy()
data2
```

```
# To convert text data into numerical values
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data2.species=label.fit_transform(data2.species)
data2.island=label.fit_transform(data2.island)
data2.sex=label.fit_transform(data2.sex)
```

```
data2
```

```
#splitting data into x and y
x=data2.iloc[:,:-1]
y=data2.iloc[:,-1]
```

```
x
```

```
y
```

```
#importing the model
from keras.models import Sequential
```

```
from keras.layers import Dense # To get layers
```

```
model=Sequential()# intialising the model
```

```
model.add(Dense(units=12,input_dim=6, activation ='tanh'))# first layer
model.add(Dense(units=12,activation='tanh'))  # second layer or hidden layer
model.add(Dense(units=1,activation='swish'))#last layer or output layer
```

```
#compiling the model
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```
model.fit(x, y , epochs=150 , batch_size=10 , verbose=1)
```
